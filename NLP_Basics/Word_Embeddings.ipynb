```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
vocab_size=10000
```

### One Hot 

```python
one_hot_rep=[one_hot(words,vocab_size) for words in sent]
```

```python
print(one_hot_rep)
```

### Word Embedding 

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embed_doc=pad_sequences(one_hot_rep,maxlen=sent_length,padding='pre')
embed_doc
```

```python
dim=15
```

```python
model=Sequential()
model.add(Embedding(vocab_size,10,input_length=sent_length))
model.compile('adam','mse')
model.summary()
```

```python
print(model.predict(embed_doc))
```
